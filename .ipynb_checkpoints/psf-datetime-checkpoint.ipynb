```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import psfws

plt.rcParams['font.size']=15
plt.rcParams['axes.labelsize']=15
plt.rcParams['axes.titlesize']=15
```

```python
def get_summaries(path):
    f_psfws = ["size_summary_psfws_df.p", "e1_summary_psfws_df.p", "e2_summary_psfws_df.p","atm_summary_psfws_df.p"]
    size, e1, e2, atm = [pickle.load(open(path+f_psfws[i], "rb")) for i in range(4)]
    return size,e1,e2,atm

def add_datetime(df):
    seed = 771963
    ws = psfws.ParameterGenerator(seed=seed)
    data_fa = ws.__getattribute__("data_fa")

    time = [data_fa.index[i] for i in range(len(data_fa))]
    speeds = {}
    for i in range(len(data_fa)):
        speeds[time[i]] = ws.get_parameters(time[i],nl=6, location="com")['speed']

    dt = []
    for i in range(len(df)):
        for j in range(len(time)):
            if all(df["speed"].iloc[i] == speeds[time[j]]):
                dt.append(time[j])
    df['month'] = [dt[i].month for i in range(len(dt))]
    df['day'] = [dt[i].day for i in range(len(dt))]
    df['hour'] = [dt[i].hour for i in range(len(dt))]
    return dt,df

def plotStepHist(ax,para,binn,change_axis=False, lbl=""):
    """plot step histogram """
    y, bin_edges, _ = ax.hist(para, bins=binn, histtype='step', color="salmon")
    bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
    if change_axis:
        bin_centers = [x+90 if x<90 else x-90 for x in bin_centers]
    return ax

def plotErrHist(ax,para,clr,binn,change_axis=False,lbl=""):
    """plot histogram of para as points with poisson uncertainty"""
    y, bin_edges, _ = ax.hist(para, bins=binn, histtype='step', color="white")
    bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
    if change_axis:
        bin_centers = [x+90 if x<90 else x-90 for x in bin_centers]
    ax.errorbar(
        bin_centers,
        y,
        yerr = y**0.5,
        marker = '.', drawstyle="steps-mid",
        linestyle ="",capsize=3, color=clr, label=lbl)
    ax.legend()
    return ax


def plot_atm_dist(dt):
    """plot distribution of month, day, hour"""
    fig,ax = plt.subplots(1,3,figsize=(15,4))
    ax[0] = plotStepHist(ax[0], [dt[i].month for i in range(len(dt))],np.arange(1,13))
    ax[0].set_xticks(np.arange(1,13,2))
    ax[0].set_xlabel("Month")

    ax[1] = plotStepHist(ax[1], [dt[i].day for i in range(len(dt))],np.arange(1,32))
    ax[1].set_xticks(np.arange(1,32,5))
    ax[1].set_xlabel("Date")

    ax[2] = plotStepHist(ax[2], [dt[i].hour for i in range(len(dt))],np.arange(0,25))
    ax[2].set_xticks(np.arange(0,25,5))
    ax[2].set_xlabel("Hour (UTC)")

    fig.suptitle("Date time distribution of psfws simulations")
    fig.tight_layout();
    

    
```

```python
path = "summariesCircle/"
size,e1,e2,atm = get_summaries(path)

dt, atm = add_datetime(atm)
plot_atm_dist(dt)
```

```python
def plot_angles_mo():
    fig, ax= plt.subplots(3,6, figsize=(20,13),sharex=True, sharey=True)
    for i in range(5,11):
        mo_seed = atm[atm['month']==i].index
        mo_angles = size.loc[mo_seed]['2p_dir']
        plotErrHist(ax[0,i-5],mo_angles,"salmon",18,change_axis=True,lbl="size")

        mo_seed = atm[atm['month']==i].index
        mo_angles = e1.loc[mo_seed]['2p_dir']
        plotErrHist(ax[1,i-5],mo_angles,"darkseagreen",18,change_axis=True, lbl="e1")

        mo_seed = atm[atm['month']==i].index
        mo_angles = e2.loc[mo_seed]['2p_dir']
        plotErrHist(ax[2,i-5],mo_angles,"steelblue",18,change_axis=True,lbl="e2")
        
        [ax[j,i-5].set_xticks(np.linspace(0,180,5)) for j in range(3)]
        [ax[j,i-5].set_xticklabels([180,135,0,45,90]) for j in range(3)]
        
        ax[2,i-5].set_xlabel("2pcf angles")
        ax[0,i-5].set_title(f"Month={i}")
        
        
    [ax[j,0].set_ylabel("Frequency") for j in range(3)]

    
def plot_angles_hr():
    fig, ax= plt.subplots(3,3, figsize=(20,12),sharex=True, sharey=True)
    hrs=[0,6,18]
    for i in range(3):
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = size.loc[hr_seed]['2p_dir']
        plotErrHist(ax[0,i],hr_angles,"salmon",18,change_axis=True,lbl="size")
        
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = e1.loc[hr_seed]['2p_dir']
        plotErrHist(ax[1,i],hr_angles,"darkseagreen",18,change_axis=True, lbl="e1")
        
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = e2.loc[hr_seed]['2p_dir']
        plotErrHist(ax[2,i],hr_angles,"steelblue",18,change_axis=True, lbl="e2")
        
        [ax[j,i].set_xticks(np.linspace(0,180,5)) for j in range(3)]
        [ax[j,i].set_xticklabels([180,135,0,45,90]) for j in range(3)]
        
        ax[2,i].set_xlabel("2pcf angles")
        ax[0,i].set_title(f"Hour={hrs[i]}")
    [ax[j,0].set_ylabel("Frequency") for j in range(3)]

    
    
def plot_autocorr():
    fig, ax = plt.subplots(1,3,figsize=(12,4),sharex=True, sharey=True)
    para = [size,e1,e2]
    hrs=[0,6,18]
    for j in range(3):
        for i in range(5,11):
            mo_seed = para[j][atm['month']==i].index
            mo_autocorr = para[j].loc[mo_seed]['autocorr']
            ax[j].scatter(i,np.mean(mo_autocorr),c="salmon", linewidth=1)
            ax[j].set_xticks([5,6,7,8,9,10])
            ax[j].ticklabel_format(axis="y",scilimits=(0,0))
            ax[j].set_title(['size',"e1","e2"][j])
    fig.suptitle("2pcf mean autocorrelation, by month")
    fig.tight_layout()

    fig, ax = plt.subplots(1,3,figsize=(12,4),sharex=True, sharey=True)
    para = [size,e1,e2]
    for j in range(3):
        for i in range(3):
            mo_seed = para[j][atm['hour']==hrs[i]].index
            mo_autocorr = para[j].loc[mo_seed]['autocorr']
            ax[j].scatter(hrs[i],np.mean(mo_autocorr),c="salmon", linewidth=1)
            ax[j].set_xticks([0,3,6,12,18])
            ax[j].ticklabel_format(axis="y",scilimits=(0,0))
            ax[j].set_title(['size',"e1","e2"][j])
    fig.suptitle("2pcf mean autocorrelation, by hour")
    fig.tight_layout()
```

```python
plot_angles_mo()
```

```python
plot_angles_hr()
```

```python
plot_autocorr()
```

```python
def plot_corrlen_mo():
    fig, ax= plt.subplots(3,6, figsize=(20,13),sharex=True, sharey=True)
    for i in range(5,11):
        mo_seed = atm[atm['month']==i].index
        mo_angles = size.loc[mo_seed]['2p_sig']
        plotErrHist(ax[0,i-5],mo_angles,"salmon",np.linspace(0,8,40),lbl="size")

        mo_seed = atm[atm['month']==i].index
        mo_angles = e1.loc[mo_seed]['2p_sig']
        plotErrHist(ax[1,i-5],mo_angles,"darkseagreen",np.linspace(0,8,40),lbl="e1")

        mo_seed = atm[atm['month']==i].index
        mo_angles = e2.loc[mo_seed]['2p_sig']
        plotErrHist(ax[2,i-5],mo_angles,"steelblue",np.linspace(0,8,40),lbl="e2")
        
        ax[2,i-5].set_xlabel("2pcf correlation length")
        ax[0,i-5].set_title(f"Month={i}")
        
        
    [ax[j,0].set_ylabel("Frequency") for j in range(3)]

    
def plot_corrlen_hr():
    fig, ax= plt.subplots(3,3, figsize=(20,12),sharex=True, sharey=True)
    hrs=[0,6,18]
    for i in range(3):
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = size.loc[hr_seed]['2p_sig']
        plotErrHist(ax[0,i],hr_angles,"salmon",np.linspace(0,8,40),lbl="size")
        
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = e1.loc[hr_seed]['2p_sig']
        plotErrHist(ax[1,i],hr_angles,"darkseagreen",np.linspace(0,8,40), lbl="e1")
        
        hr_seed = atm[atm['hour']==hrs[i]].index
        hr_angles = e2.loc[hr_seed]['2p_sig']
        plotErrHist(ax[2,i],hr_angles,"steelblue",np.linspace(0,8,40), lbl="e2")
        
        ax[2,i].set_xlabel("2pcf correlation length")
        ax[0,i].set_title(f"Hour={hrs[i]}")
    [ax[j,0].set_ylabel("Frequency") for j in range(3)]
```

```python
plot_corrlen_mo()
```

```python
plot_corrlen_hr()
```
