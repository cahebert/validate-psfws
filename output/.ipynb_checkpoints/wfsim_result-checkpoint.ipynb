```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

def plotstars(outfile):
    with open(outfile, 'rb') as f:
        data = pickle.load(f)

    for key in data.keys():
        data[key] = np.array(data[key])

    (thx, thy, seed, x, y, sigma, e1, e2, args, atmSummary, atmKwargs) = tuple(data.values())
    idk = np.hypot(e1,e2)
    fig, ax = plt.subplots()
    ax.quiver(x,y,e1,e2,idk,pivot="mid")
```

```python
plotstars("out.pkl")
```

```python
plotstars("out1.pkl")
```

```python
plotstars("out2.pkl")
```
